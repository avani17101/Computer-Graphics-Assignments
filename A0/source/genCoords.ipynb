### Decagonal prism

```python
# gen coords
vert = [['0.0f','-a'],
        ['a*cos(54*pi/180)','-a*sin(54*pi/180)'],
        ['a*cos(18*pi/180)', '-a*sin(18*pi/180)'], 
        ['a*cos(18*pi/180)', 'a*sin(18*pi/180)'],
        ['a*cos(54*pi/180)', 'a*sin(54*pi/180)'],
        ['0.0f','a'],
        
        ['-a*cos(54*pi/180)', 'a*sin(54*pi/180)'],
        ['-a*cos(18*pi/180)', 'a*sin(18*pi/180)'],
        ['-a*cos(18*pi/180)', '-a*sin(18*pi/180)'], 
        ['-a*cos(54*pi/180)','-a*sin(54*pi/180)'],
        ['0.0f','-a']]
       

print('//side coords')
for i in range(1,len(vert)):
    v = vert[i]
    v_ = vert[i-1]
    color = [round(random.uniform(0, 1),1),round(random.uniform(0, 1),1),round(random.uniform(0, 1),1)]
    
    print('glColor3f('+ str(color[0])+'f, '+ str(color[1])+'f, '+ str(color[2])+'f );') 
    print('glVertex3f('+v_[1]+ ' ,0.0f, '+v_[0]+'); \n')
    
    print('glColor3f('+ str(color[0])+'f, '+ str(color[1])+'f, '+ str(color[2])+'f );') 
    print('glVertex3f('+v[1]+ ' ,0.0f, '+v[0]+'); \n')
    
    print('glColor3f('+ str(color[0])+'f, '+ str(color[1])+'f, '+ str(color[2])+'f );') 
    print('glVertex3f('+v[1]+ ' ,h, '+v[0]+'); \n')
    
    print('glColor3f('+ str(color[0])+'f, '+ str(color[1])+'f, '+ str(color[2])+'f );') 
    print('glVertex3f('+v_[1]+ ' ,h, '+v_[0]+'); \n')
    

    
print("//bottom face")
for i in range(1,len(vert)):
    v = vert[i]
    v_ = vert[i-1]
    
    print('glColor3f(0.0f, 0.5f, 0.0f);') 
    print('glVertex3f('+v_[1]+ ' ,0.0f, '+v_[0]+'); \n')
    
    print('glColor3f(0.0f, 0.5f, 0.0f);') 
    print('glVertex3f('+v[1]+ ' ,0.0f, '+v[0]+'); \n')
    
    print('glColor3f(0.0f, 0.5f, 0.0f);') 
    print('glVertex3f(0.0f , 0.0f, 0.0f); \n')
    
    print('glColor3f(0.0f, 0.5f, 0.0f);') 
    print('glVertex3f('+v_[1]+ ' ,0.0f, '+v_[0]+'); \n \n')
    
print('//top face')
for i in range(1,len(vert)):
    v = vert[i]
    v_ = vert[i-1]

    print('glColor3f(0.0f, 0.0f, 0.2f);') 
    print('glVertex3f('+v_[1]+ ' ,h, '+v_[0]+'); \n')
    
    print('glColor3f(0.0f, 0.0f, 0.2f);') 
    print('glVertex3f('+v[1]+ ' ,h, '+v[0]+'); \n')
    
    print('glColor3f(0.0f, 0.0f, 0.2f);') 
    print('glVertex3f(0.0f , h, 0.0f); \n')
    
    print('glColor3f(0.0f, 0.0f, 0.2f);') 
    print('glVertex3f('+v_[1]+ ' ,h, '+v_[0]+'); \n \n')
```
